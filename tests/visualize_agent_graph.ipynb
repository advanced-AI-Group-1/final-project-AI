# LangGraph 에이전트 그래프 시각화

이 노트북은 ReportAgent의 LangGraph 워크플로우를 시각화합니다.

```python
# 필요한 라이브러리 임포트
import sys
from pathlib import Path

from dotenv import load_dotenv

# 프로젝트 루트 디렉토리를 sys.path에 추가
project_root = Path().resolve().parent
sys.path.append(str(project_root))

# 환경 변수 로드
load_dotenv()
```

```python
# ReportAgent 클래스 임포트
from app.domain.report_generation.agent import ReportAgent
from app.domain.report_generation.prompts import DEFAULT_SECTIONS

# ReportAgent 인스턴스 생성
agent = ReportAgent()
```

```python
# 워크플로우 생성 및 시각화
workflow = agent._create_workflow()
graph = workflow.compile()

# 그래프 시각화
graph_html = graph.write_html()
from IPython.display import HTML
HTML(graph_html)
```

## 그래프 구조 분석

### 노드 정보

```python
# 노드 정보 출력
print("=== 노드 정보 ===")
for node in workflow.nodes:
  print(f"- {node}")
```

### 엣지 정보

```python
# 엣지 정보 출력
print("=== 엣지 정보 ===")
for edge in workflow.edges:
  print(f"- {edge[0]} → {edge[1]}")

# 조건부 엣지 정보 출력
print("\n=== 조건부 엣지 정보 ===")
for node, edges in workflow.conditional_edges.items():
  print(f"- {node} 노드의 조건부 엣지:")
  for condition, target in edges.items():
    print(f"  - 조건 '{condition}' 일 때 → {target}")
```

### 섹션 정보

```python
# 섹션 정보 출력
print("=== 섹션 정보 ===")
for i, section in enumerate(DEFAULT_SECTIONS):
  print(f"{i + 1}. {section['name']}: {section['description']}")
```
